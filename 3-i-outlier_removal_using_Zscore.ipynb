```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("loan.csv")
```

```python
# Boxplot se hm Minimum aur maximum values find kr skty isisliye outliers remove krny kaliye IQR use krty , 
# Agr normal distribution curve hy tou waha Zscore se check krty
```

```python
# z = x-u/std
# z Range = -3 to +3 
# data beyond this is considered as outlier
#  we   can find range by subtracting mean from 3rd standard deviation
```

```python
sns.boxplot(x = "CoapplicantIncome", data= df)
plt.show()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
min_range = df["CoapplicantIncome"].mean() - (3*df["CoapplicantIncome"].std())
max_range = df["CoapplicantIncome"].mean() + (3*df["CoapplicantIncome"].std())
min_range, max_range
```

```python
new_data1 = df[df["CoapplicantIncome"]<=max_range]
```

```python
new_data1.shape
```

```python
sns.boxplot(x = "CoapplicantIncome", data = new_data1)
plt.show()
```

### Z-Score

```python
z_score = (df["CoapplicantIncome"]-df["CoapplicantIncome"].mean())/df["CoapplicantIncome"].std()
```

```python
df["z_score"] = z_score
```

```python
df[df["z_score"]<=3]
```
