### Ordinal Encoding is performed by two Ways , One is Scikit learn and the other is Map Function


```python
import  pandas as pd
df = pd.DataFrame({"Size":["s", "m", "l", "xl", "s", "m", "l", "s", "s", "l", "xl", "m"]})
df
```

```python
order_data = [["s", "m", "l", "xl"]]
from sklearn.preprocessing import  OrdinalEncoder
oe  = OrdinalEncoder(categories= order_data) # categories = auto  ho tou alphabetically order krta

```

```python
oe.fit(df[["Size"]])
```

```python
df["size_encode"] = oe.transform(df[["Size"]])
```

```python
df
```

###### Map function

```python
order_data1 = {"s" : 5, "m":10, "l": 15, "xl":20}
df["size_encode1"]=df["Size"].map(order_data1)
```

```python
df
```

```python
df = pd.read_csv("loan.csv")
```

```python
df["Property_Area"].unique()
```

```python
en_data_order = [['Urban', 'Rural', 'Semiurban']]
```

```python
oen = OrdinalEncoder(categories=en_data_order)
df["Property_Area"] =oen.fit_transform(df[["Property_Area"]])
```

```python
df["Married"].unique()
```

```python
df["Married"] = df["Married"].fillna(df["Married"].mode()[0])
```

```python
order_data2 = [['No', 'Yes']]
```

```python
oen1 = OrdinalEncoder(categories=order_data2)
df["Married"] = oen1.fit_transform(df[["Married"]])
```
