##### Transformation : NoN - Normal distributed data ko normal distributed me convert krna 

###### Function Transformation can be used to change the pattern of data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df  = pd.read_csv("loan.csv")
```

```python
df.isnull().sum()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

##### Remove Outlier

```python
q1 = df["CoapplicantIncome"].quantile(0.25)
```

```python
q3 = df["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)
```

```python
min_range, max_range
```

```python
df= df[df["CoapplicantIncome"]<=max_range]
```

```python
df.info()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func = np.log1p) #agr data me 0 ho ga tou tb b log1p work kry ga , log10, loge nhi krta , yw infinity dety
```

```python
ft.fit(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_ft"] = ft.transform(df[["CoapplicantIncome"]])
```

```python
plt.subplot(1, 2,1)
sns.distplot(df["CoapplicantIncome"])
plt.subplot(1, 2,2)
sns.distplot(df["CoapplicantIncome_ft"])
plt.show()
```

#### We can use transform fuction without removing the outliers

```python
df1  = pd.read_csv("loan.csv")
```

```python
sns.distplot(df1["CoapplicantIncome"])
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer
```

```python
ft1 = FunctionTransformer(func = np.log1p)
```

```python
ft1.fit(df1[["CoapplicantIncome"]])
```

```python
df1["CoapplicantIncome_ft"] = ft1.transform(df1[["CoapplicantIncome"]])
```

```python
plt.subplot(1, 2,1)
sns.distplot(df1["CoapplicantIncome"])
plt.subplot(1, 2,2)
sns.distplot(df1["CoapplicantIncome_ft"])
plt.show()
```

##### it causes no affect

#### We can use other functions ,can also customize for fucntion transform


```python
ft2 = FunctionTransformer(func = lambda x: x**2)
ft2.fit(df1[["CoapplicantIncome"]])
```

```python
df1["CoapplicantIncome_ft"] = ft2.transform(df1[["CoapplicantIncome"]])
```

```python
plt.subplot(1, 2,1)
sns.distplot(df1["CoapplicantIncome"])
plt.subplot(1, 2,2)
sns.distplot(df1["CoapplicantIncome_ft"])
plt.show()
```
